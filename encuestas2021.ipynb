```python
# word cloud with open questions
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import wikipedia

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# pruebas
page = wikipedia.page("software")
text = page.content

# Generar el wordcloud
stopwords = set(STOPWORDS)
stopwords.update(["stopword1", "stopword2"])

cloud = WordCloud(stopwords=stopwords).generate(text)
plt.figure(figsize=(18,18), facecolor=None )
plt.imshow(cloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
#cloud.to_file("wordcloud.png")
```

```python
# Librerias necesarias
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

```python
# Analisis respuesta 5 de primaria

# Procesar el texto para eliminar las stopwords
spanish_sw = stopwords.words("spanish")

nombre_encuesta = 'encuestaPq5.txt'

dirname = path.dirname(nombre_encuesta)

# Read the whole text
text = open(path.join(dirname, nombre_encuesta)).read()


cloud = WordCloud(stopwords=spanish_sw).generate(text)
plt.figure(figsize=(15,15), facecolor=None )
plt.imshow(cloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

```
